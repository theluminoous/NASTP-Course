## 2. Histograms
---
Histograms are used to visualize the <b>distribution of a dataset</b> by <b>grouping data into bins</b> and showing how many values fall into each bin.   
It helps identify patterns like <b>skewness, spread, and outliers</b> in numerical data.

```python
# used when we need to show info in bars
# to create data using random, we have to import numpy and random
import numpy as np
import random
```

```python
np.random.seed(12)
ai_st_age = np.random.randint(18,45,100)
bd_st_age = np.random.randint(18,45,100)
```

```python
print(ai_st_age)
print(bd_st_age)
```

```python
bd_st_age
```

```python
# hist() is used to creat histogram
plt.hist(ai_st_age)
plt.title('AI Student\'s age')
plt.xlabel('Students age category')
plt.ylabel('No. of students')
plt.show()
```

```python
bins = [15,20,25,30,35,40,45] #used to make groups
plt.hist(ai_st_age, bins)
plt.title('AI Student\'s age')
plt.xlabel('Students age category')
plt.ylabel('No. of students')
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ai_st_age, bins, rwidth=0.8) #to set width
plt.title('AI Student\'s age')
plt.xlabel('Students age category')
plt.ylabel('No. of students')
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ai_st_age, bins, rwidth=0.8, histtype = 'step') # to change hist style
# plt.hist(ai_st_age, bins, rwidth=0.8, histtype = 'bar')
# plt.hist(ai_st_age, bins, rwidth=0.8, histtype = 'stepfilled')

plt.title('AI Student\'s age')
plt.xlabel('Students age category')
plt.ylabel('No. of students')
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ai_st_age, bins, rwidth=0.8, orientation='horizontal') # to change orientation
plt.title('AI Student\'s age')
plt.xlabel('Students age category')
plt.ylabel('No. of students')
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ai_st_age, bins, rwidth=0.8, color='r', label='AI students')
plt.hist(bd_st_age, bins, rwidth=0.8, color='g', label='BD students')
plt.title('AI Student\'s age')
plt.xlabel('Students age category')
plt.ylabel('No. of students')
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist([ai_st_age, bd_st_age], bins, rwidth=0.8, color=['r','g'], label=['AI students', 'BD students'])
plt.title('AI Student\'s age')
plt.xlabel('Students age category')
plt.ylabel('No. of students')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (16,9)) # to change fig size 

bins = [15,20,25,30,35,40,45]
plt.hist([ai_st_age, bd_st_age], bins, rwidth=0.8, color=['r','g'], label=['AI students', 'BD students'])
plt.title('AI Student\'s age')
plt.xlabel('Students age category')
plt.ylabel('No. of students')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (10,4)) # to change fig size 

bins = [15,20,25,30,35,40,45]
plt.hist([ai_st_age, bd_st_age], bins, rwidth=0.8, color=['r','g'], label=['AI students', 'BD students'])
plt.title('AI Student\'s age')
plt.xlabel('Students age category')
plt.ylabel('No. of students')
plt.legend()
plt.grid()
plt.show()
```

## 3. Bar Chart
---
Bar Chart is used to visually compare <b>discrete categories or groups based on their values</b>.   
It is ideal for showing counts, frequencies, or other measures across distinct labels like names, months, or product types.

```python
classes = ['Python','R','AI','ML','DS']
class1_students = [30,10,20,25,10]
class2_students = [40,5,20,20,10]
class3_students = [35,5,30,15,15]
```

```python
# bar() is used to create bar chart
plt.bar(classes, class1_students)
plt.show()
```

```python
plt.barh(classes, class1_students) #to plot horizantaly
plt.show()
```

```python
plt.figure(figsize=(8,3), dpi=100, facecolor='g', frameon=True) #fig parameters
plt.bar(classes, class1_students, width=0.6, align='edge', color='r', 
        edgecolor='g', linewidth=3, alpha=0.7, linestyle='-.',
       label='Cass 1 Students')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,3), dpi=100, facecolor='#f5f5f5', frameon=True) #fig parameters
plt.bar(
    classes,
    class1_students, 
    width=0.6, 
    align='edge', 
    color='#000080',         # Slate Blue 
    edgecolor='#000000',     # Sea Green
    linewidth=3, 
    alpha=0.7, 
    linestyle='-.',
    label='Cass 1 Students'
)

plt.legend()
plt.title("Number of Students per Class", fontsize=14)
plt.xlabel("Classes", fontsize=12)
plt.ylabel("Number of Students", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.savefig('bar_chart_of_classes.png', dpi=1000, bbox_inches='tight') # to save chart
plt.show()
```

```python
plt.barh(
    classes, 
    class1_students, 
    align='center', 
    color='#000080',         # Slate Blue 
    edgecolor='#000000',     # Sea Green
    linewidth=3, 
    alpha=0.7, 
    linestyle='-.',
    label='Cass 1 Students'
) #barh don't need width
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,3))

plt.bar(classes, class1_students, width=0.3,color='r',label='Cass 1 Students')
plt.bar(classes, class2_students, width=0.3,color='g',label='Cass 2 Students')

plt.title('Bar Chart of Classes', fontsize = 12)
plt.xlabel('Classes', fontsize = 10)
plt.ylabel('No. of Students', fontsize = 10)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,3))

import numpy as np
classes_index = np.arange(len(classes))
print(classes_index)

# ['Python','R','AI','ML','DS'] --> [0 1 2 3 4]

width = 0.3
plt.bar(classes_index, class1_students, width, color='r',label='Class 1 Students')
plt.bar(classes_index + width, class2_students, width,color='g',label='Class 2 Students')
plt.bar(classes_index + width + width, class3_students, width,color='b',label='Class 3 Students')

plt.title('Bar Chart of Classes', fontsize = 22)
plt.xlabel('Classes', fontsize = 20)
plt.ylabel('No. of Students', fontsize = 20)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,4))

import numpy as np
classes_index = np.arange(len(classes))

# ['Python','R','AI','ML','DS'] --> [0 1 2 3 4]

width = 0.3
plt.bar(classes_index, class1_students, width, color='r',label='Class 1 Students')
plt.bar(classes_index + width, class2_students, width,color='g',label='Class 2 Students')
plt.bar(classes_index + width + width, class3_students, width,color='b',label='Class 3 Students')

plt.xticks(
    classes_index + width, #location to show labels
    classes,               #Labels to show
    rotation = 10           #to rotate labels
) #to change xlabel


plt.title('Bar Chart of Classes', fontsize = 12)
plt.xlabel('Classes', fontsize = 10)
plt.ylabel('No. of Students', fontsize = 10)
plt.legend()
plt.show()
```

```python
# to plot horizantally

plt.figure(figsize=(16,9))
width = 0.2
plt.barh(classes_index, class1_students, width,color='r',label='Cass 1 Students')
plt.barh(classes_index + width, class2_students, width,color='g',label='Cass 2 Students')
plt.barh(classes_index + width + width, class3_students, width,color='m',label='Cass 3 Students')

plt.yticks(classes_index + width, classes) # to change label place

plt.title('Bar Chart of Classes', fontsize = 22)
plt.ylabel('Classes', fontsize = 20)
plt.xlabel('No. of Students', fontsize = 20)
plt.legend()
plt.show()
```
